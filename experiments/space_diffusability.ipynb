```python
# Start Generation Here
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the two CSVs
file_v11 = "./reconstruction_loss-diffusion-rocstories-16-d=5-v1.1.csv"
file_v10 = "./reconstruction_loss-diffusion-rocstories-16-d=5-v1.0.csv"
df_v11 = pd.read_csv(file_v11)
df_v10 = pd.read_csv(file_v10)

# Set up Seaborn style
sns.set(style="whitegrid", font_scale=1.2)

# Create the plot
plt.figure(figsize=(10, 6), dpi=300)
plt.plot(df_v11["t"], df_v11["loss_x_0"], label="v1.1", color="tab:blue", linewidth=2)
plt.plot(df_v10["t"], df_v10["loss_x_0"], label="v1.0", color="tab:orange", linewidth=2, linestyle="--")

# Labels, title, legend
plt.xlabel("t")
plt.ylabel("loss_x_0")
plt.title("Reconstruction Loss vs t for ROCStories (d=5)")
plt.legend(title="Version")
plt.tight_layout()

# Display
plt.show()
```
